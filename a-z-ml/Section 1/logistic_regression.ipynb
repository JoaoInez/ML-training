```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
from visualize_classifier import visualize_classifier
%matplotlib inline
```

```python
X = np.array([
    [3.1, 7.2],
    [4, 6.7],
    [2.9, 8],
    [5.1, 4.5],
    [6, 5],
    [5.6, 5],
    [3.3, 0.4],
    [3.9, 0.9],
    [2.8, 1],
    [0.5, 3.4],
    [1, 4],
    [0.6, 4.9]
])
y = np.array([0, 0, 0, 1, 1, 1, 2, 2, 2, 3, 3, 3])
```

```python
model = LogisticRegression(solver="liblinear", C=1)

model.fit(X, y)
```

```python
predictions = model.predict(X)
predictions
```

```python
sns.scatterplot(x=X[:, 0], y=X[:, 1], hue=predictions, palette="deep", marker="+", s=200)
sns.scatterplot(x=X[:, 0], y=X[:, 1], hue=y, palette="deep")
```
