```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.neighbors import KNeighborsClassifier
%matplotlib inline
```

```python
data = np.loadtxt("../input/data.txt", delimiter=",")
data[:5]
```

```python
X = data[:, :-1]
y = data[:, -1]
```

```python
sns.scatterplot(X[:, 0], X[:, 1], hue=y)
```

```python
num_neighbors = 12

model = KNeighborsClassifier(num_neighbors, weights="distance")
model.fit(X, y)
```

```python
test_datapoint = [5.1, 3.6]

_, indices = model.kneighbors([test_datapoint])
indices
```

```python
sns.scatterplot(X[:, 0], X[:, 1], hue=y)
sns.scatterplot(X[indices][0][:, 0], X[indices][0][:, 1], s=100)
sns.scatterplot([test_datapoint[0]], [test_datapoint[1]], s=100, marker="X")
```

```python
model.predict([test_datapoint])[0]
```
