```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
%matplotlib inline
```

```python
X = np.loadtxt("../../input/data_clustering.txt", delimiter=",")
X[:5]
```

```python
sns.scatterplot(x=X[:, 0], y=X[:, 1])
```

```python
model = KMeans(init='k-means++', n_clusters=5, n_init=10)
model.fit(X)
```

```python
sns.scatterplot(x=X[:, 0], y=X[:, 1], hue=model.labels_, palette="deep")
sns.scatterplot(x=model.cluster_centers_[:, 0], y=model.cluster_centers_[:, 1], color="0", s=100, marker="s")
```
