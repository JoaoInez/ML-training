```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.neighbors import NearestNeighbors
%matplotlib inline
```

```python
X = np.array([
    [2.1, 1.3],
    [1.3, 3.2],
    [2.9, 2.5],
    [2.7, 5.4],
    [3.8, 0.9],
    [7.3, 2.1],
    [4.2, 6.5],
    [3.8, 3.7],
    [2.5, 4.1],
    [3.4, 1.9],
    [5.7, 3.5],
    [6.1, 4.3],
    [5.1, 2.2],
    [6.2, 1.1]
])
```

```python
sns.scatterplot(x=X[:, 0], y=X[:, 1])
```

```python
k = 5

model = NearestNeighbors(n_neighbors=k, algorithm='ball_tree').fit(X)
model
```

```python
test_datapoint = [4.3, 2.7]

distances, indices = model.kneighbors([test_datapoint])
indices
```

```python
print("\nK Nearest Neighbors:")
for rank, index in enumerate(indices[0][:k], start=1):
    print(str(rank) + " ==>", X[index])
```

```python
sns.scatterplot(X[:, 0], X[:, 1])
sns.scatterplot(X[indices][0][:, 0], X[indices][0][:, 1], s=100)
sns.scatterplot([test_datapoint[0]], [test_datapoint[1]], marker="X", s=100)
```
